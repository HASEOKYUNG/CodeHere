## Move Kernel
CMD에서<br>
1)`c:`를 입력해 드라이브로 이동하고<br>
2)`cd 폴더명`를 입력해 폴더로 이동한 뒤<br>
3)`jupyter notebook`으로 실행한다.<br>
  $\;\;\;$ py파일을 실행시키고 싶을 땐 `python 파일명.py`로 한다.

## jupyter notebook key
- [H]: jupyter notebook 단축 key 목록 출력
- [A], [B]: 위, 아래로 cell 추가
- `[ctrl]+[F]`, `[F]`: 검색, replace 기능
- [shift]+[K], [shift]+[J], `[ctrl]+[A]`: 위, 아래 cell 추가 선택, 전체 cell 선택
- [C], [shift]+[V], [V]: 복사, 현재 cell의 위, 아래로 붙여넣기
- [D]+[D], [X]: 선택 cell 삭제
- [Z]: 삭제한 cell 복구
- [shift]+[Enter], [Alt]+[Enter]: cell 돌리기, cell 돌리고 아래에 cell 추가
- [i]+[i]: cell 돌리기 중단
- [Enter], [Esc]: 파란 모드에서 초록 모드로, 초록 모드에서 파란 모드로 선택    
- `[shift]+[M]`, `[ctrl]+[shift]+[-]`: cell 병합, cell 분리
- [Tab], [shift]+[Tab]: 앞, 뒤로 Tab 이동
- [shift]+[PgUp], [shift]+[PgOn], [ctrl]+[A]: 현재 cell line의 위, 아래 line 추가선택, cell 내 전체 선택
- [shift]+[Tab]: 함수, Method API 정보 출력
- [ctrl]+[D]: 선택 cell line 삭제
- [ctrl]+[Z]: cell 수정사항 복원
- `[ctrl]+[/]`: 선택 cell line 주석처리
- [ctrl]+[Delete], [ctrl]+[End]: 선택 cell line 공백 지우기, 한 단어씩 옆으로 이동  
- [S], [ctrl]+[S]: file 저장
- `[Y]`: code 모드로 cell 전환
- `[M]`, [1], [2], [3], [4], [5], [6]: markdown 모드로 cell 전환(숫자가 클수록 글씨가 작게 설정됨)
- [shift]+[L], [L]: cell 내 line number 설정, 출력/숨기기
- [0]+[0]: file 연결 끊었다가 다시 시작

## Import
- Module 전체를 불러온다. 이때, Module.Function()식으로 모듈명을 붙여 사용해야 한다.<br>
  Package 전체를 부르는 방법도 같은 방법이다.
```python
import 모듈명                        
```
- Module의 특정 함수만을 불러온다. 동시에 여러 함수를 부르는 것도 가능하다. 이때, Function()식으로, `모듈명을 붙이지 않고 사용한다.`<br>
  Package의 특정 Module만, Library의 특정 Package, Module만 불러오는 방법도 같은 방법이며 `패키지명, 라이브러리명을 붙이지 않고 사용한다.`
```python
from 모듈명 import 함수         
from 패키지명.모듈명 import 함수  
from 모듈명 import 함수1, 함수2, ...
```

## LaTeX
LaTeX를 조정하는 방법은 아래와 같다.
- Size
  - $\scriptsize \text{scriptsize}$
  - $\small \text{small}$
  - $\normalsize \text{normalsize}$
  - $\large \text{large}$
  - $\Large \text{Large}$
- Blank
  - $a \, b$
  - $a ~ b$
  - $a \quad b$
  - $a \qquad b$
- Sort
  - $$로 열고 닫으면 LaTeX는 가운데 정렬된다. 
- Color
  - 전체)$\color{Red}{x^2}$, $\color{Green}{x^2}$, $\color{Blue}{x^2}$,  $\color{Navy}{x^2}$, $\color{Purple}{x^2}$
  - 일부)${\color{Red}x^2}$, $x^{\color{Red}{2}}$
  
자주 쓰는 기호는 아래와 같다.
- Params
  - $\alpha$, $\beta$, $\gamma$, $\delta$, $\epsilon$, $\lambda$ 
  - $\mu$, $\sigma$
  - $\theta$, $\tau$, $\nu$, $\phi$, $\psi$
  - $\chi$, $\Gamma$, $\Delta$, $\Phi$
  - $\rho$, $\upsilon$, $\omega$, $\eta$, $\Omega$, $\Lambda$
- Calculate
  - $\underset{lower}{\overset{upper}\prod}$
  - $\underset{lower}{\overset{upper}\sum}$
  - $\int_{lower}^{upper}$
  - $\ln x$, $\log_e x$
  - $_nC_k$
- Math Symbol
  - $\pm$, $\times$, $\div$, $\left(\frac{a}{b}\right)$
  - $\leq$, $\geq$, $\neq$, $\fallingdotseq$
  - $\{A\}\cap\{B\}$, $\{A\}\cup\{B\}$
  - $\underset{n \to \infty}\lim x$
  - $\sqrt x, \sqrt[3]x$
  - $\therefore$, $\because$
  - $f(n) = \begin{cases} f(n_1) & if~ n_1 < 0 \\ f(n_2) & if~ n_2 > 0 \end{cases}$
  - $\infty$, $\pi$
- Text Symbol
  - $\blacktriangleright$, $\triangleright$, $\bullet$
  - $\rightarrow$, $\leftarrow$, $\longrightarrow$, $\longleftarrow$
  - $\Rightarrow$, $\Leftarrow$, $\Longrightarrow$, $\Longleftarrow$
  - $\leftrightarrow$, $\longleftrightarrow$
  - $\begin{vmatrix} & 1열 & 2열~ \\ 1행 & x_1 & x_2~ \\ 2행 & y_1 & y_2~ \end{vmatrix}$
  - $\cdot$, $\cdots$, $\ddots$
- [참고] https://jjycjnmath.tistory.com/117

## Markdown 
- `강조`, **강조**, *강조* &#160;로 일부분을 강조할 수 있다. 그러나 LaTeX를 강조할 순 없다.
- ` ```python` 입력 후 코드를 작성한 뒤 ` ``` `로 닫으면 code cell에서 작성했을 때와 똑같이 출력된다.
- `Table 작성`<br>
  아래의 주소에서 표를 작성한 뒤 깨질 경우 header에 `&#160;` 또는 `&ensp;`로 띄어쓰기 조정해 표를 작성한다.<br>
  [markdown table] https://www.tablesgenerator.com/markdown_tables
